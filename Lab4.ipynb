```python
f = pd.read_csv("heart.csv")
print(f)
```

```python
type_dict = {int('0') : str('female'), int('1') : str('male')}
f['gender'] = f['sex'].map(type_dict)
type_dict2 = {int('0') : str('has heart disease'), int('1') : str('does not have heart disease')}
f['hasHeartDisease'] = f['target'].map(type_dict2)
f.describe()
```

```python
h_d1 = f.trestbps.loc[f.hasHeartDisease == 'has heart disease']
h_d2 = f.trestbps.loc[f.hasHeartDisease == 'does not have heart disease']
mean1 = np.mean(h_d1)
std1 = np.std(h_d1)
mean2 = np.mean(h_d2)
std2 = np.std(h_d2)
print(mean1, mean2)
print(std1, std2)
```

```python
fig, axs = plt.subplots(ncols=2,sharey=True)
axs[0].boxplot(h_d1, vert = False)
axs[0].set_title('has disease')
axs[1].boxplot(h_d2, vert = False)
axs[1].set_title('does not have disease')
```

```python
fig, axs = plt.subplots(ncols=2,sharey=True)
axs[0].hist(h_d1)
axs[0].set_title('has disease')
axs[1].hist(h_d2)
axs[1].set_title('does not have disease')
```

```python
from numpy.random import randn
from numpy.random import seed
from numpy import mean
from numpy import var
from math import sqrt

def cohend(h_d1, h_d2):
    c1, c2 = len(h_d1), len(h_d2)
    d1, d2 = var(h_d1, ddof = 1), var(h_d2, ddof = 1)
    s = sqrt(((c1 - 1) * d1 + (c2 - 1) * d2) / (c1 + c2 - 2))
    return (a - b) / s

cohend(h_d1, h_d2)
```

```python
#Count the number with the disease for each gender type
hasDiseaseCount=f[f.hasHeartDisease == 'has heart disease'].groupby('gender').count().target

#Count the number of gender type
totalCount=f.groupby('gender').count()['hasHeartDisease']

#combine into a dataframe (both are indexed with gender, so will be matched) and specify the columns
p=pd.concat([hasDiseaseCount, totalCount], axis=1)
p.columns = ["heartDiseaseCount", "totalCount"]

#create a new column and calculate the proportion
p['propHeartDisease']=p["heartDiseaseCount"]/p["totalCount"]

#print the results
print(p.head())
```
