```python
import pandas as pd
import numpy as np
f = pd.read_csv("censusCrimeClean.csv")
print(f)
```

```python
import scipy
from scipy import stats
from matplotlib import pyplot as plt
a = f["medIncome"]
b = f["ViolentCrimesPerPop"]
p_value = scipy.stats.pearsonr(a,b)
s_value = scipy.stats.spearmanr(a,b)
print(p_value)
print(s_value)
f.plot.scatter(x = "medIncome", y= "ViolentCrimesPerPop", c = 'red')
plt.show()
```

```python
x_col = ["medIncome"]
y_columns = ["PctNotSpeakEnglWell", "PctLargHouseOccup", "PersPerOccupHous"]

for y_col in y_columns:
    figure = plt.figure
    ax = plt.gca()
    ax.scatter(f[x_col], f[y_col])
    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)
    ax.set_title("{} vs {}".format(x_col, y_col))
    plt.show()
```
